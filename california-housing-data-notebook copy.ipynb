```python
# Import housing dataset from sklearn
import sklearn.datasets
housing_dataset = sklearn.datasets.fetch_california_housing()
```

```python
X = housing_dataset["data"]
y = housing_dataset["target"]
```

```python
# Test-train split, 30-70
# This doesn't incorporate weights, though the sample already did this so it doesn't matter
import sklearn.model_selection
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size = 0.3)
```

```python
y_train.shape
```

```python
# Create an adaboost classifier with decision stumps for now
# TODO: Should the random state be freed?
import sklearn.tree
import sklearn.ensemble

TREE_DEPTH: int = 10
NUM_ESTIMATORS: float = 1000
LEARNING_RATE: float = 1

base_estimator = sklearn.tree.DecisionTreeRegressor(max_depth = TREE_DEPTH)

adaboost_regressor = sklearn.ensemble.AdaBoostRegressor(
                                    estimator = base_estimator,
                                    n_estimators = NUM_ESTIMATORS,
                                    learning_rate = LEARNING_RATE,
                                    random_state = 0)
```

```python
adaboost_regressor.fit(X_train, y_train)
```

```python
sklearn.metrics.mean_squared_error(adaboost_regressor.predict(X_train), y_train)
```

```python
import sklearn.metrics
y_test_pred = adaboost_regressor.predict(X_test)

sklearn.metrics.mean_squared_error(y_test_pred, y_test)
```
